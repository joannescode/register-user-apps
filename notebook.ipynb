```python
def validate_user(name, email, password):
    """Validates the user's name, email, and password.

    This fucton calls validate_name, validate_email, and validade_password from python_fuctions.py to check if the provided credentials are valid.

    Args:
        name (str): The user's name.
        email (str): The user's email.
        password (str): The user's password.

    Returns:
        bool: True if all validations pass, otherwise False.
    """
    if (
        validate_name(name)
        and validate_email(email)
        and validate_password(password)
    ):
        return True
    else:
        raise ValueError("One or more validations did not pass")

def register_user(name, email, password):
    """Register User in case the user's name, email, and password pass.

    Args:
        name (str): The user's name.
        email (str): The user's email.
        password (str): The user's password.

    Returns:
        dict: In case all validations pass, return a dict from user_registred with credentials, otherwise False.
    """
    if validate_user(name, email, password):
        user_registred = {"name": name, "email": email, "password": password}
        return user_registred
    else:
        return False
    
register_user(name="Lucas", email="datacamp_student@gmail.com", password="Datacamp#15")
```
